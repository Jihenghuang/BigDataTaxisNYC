```python
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
import requests
from shapely.geometry import Point, MultiPolygon
import time
```

## I. Loading and processing stations data

```python
url_stations = 'http://www.citibikenyc.com/stations/json'
json_stations = requests.get(url_stations).json()

stations = pd.DataFrame(json_stations['stationBeanList'])
stations.head()
```

```python
stations['geometry'] = stations.apply(lambda z: Point(z.longitude, z.latitude), axis=1)
stations = gp.GeoDataFrame(stations)
stations.head()
```

## II. Loading Zipcodes 

```python
zipcodes = gp.read_file('NYCzipcodeshapefile2.geojson')
zipcodes.head()
```

## III. Categorize stations by zipcode


```python
for i,n in enumerate(stations['geometry']):
    for x,z in enumerate(zipcodes['geometry']):
        if n.intersects(z):
            print "Station {} is in {}".format(stations['stAddress1'][i],zipcodes['postalCode'][x])
```
